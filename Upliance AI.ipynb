# Merging all 3 tables

```python

order_user_details = pd.merge(order_details, user_details, on='User ID', how='inner')

full_data = pd.merge(order_user_details, cooking_sessions, on=['User ID', 'Dish Name'], how='inner')

print(full_data.head())
```

# Visualising Popular dishes (per count)

```python
popular_dishes = full_data.groupby('Dish Name').agg(OrderCount=('Order ID', 'nunique')).reset_index()
popular_dishes_sorted = popular_dishes.sort_values(by='OrderCount', ascending=False)
print(popular_dishes_sorted)

plt.figure(figsize=(10, 6))
sns.barplot(x='OrderCount', y='Dish Name', data=popular_dishes_sorted.head(10), palette='viridis')
plt.title('Popular Dishes')
plt.xlabel('Number of Orders')
plt.ylabel('Dish Name')
plt.show()
```

```python
user_activity = full_data.groupby('User ID').agg(
    TotalCookingSessions=('Session ID_x', 'nunique'),  
    TotalOrders=('Order ID', 'nunique') 
).reset_index()
```

# Relationship Between Cooking Sessions and Orders is linear

```python
plt.figure(figsize=(12, 6), facecolor='lightgreen')
sns.barplot(x='TotalCookingSessions', y='TotalOrders', data=user_activity)
plt.title('Relationship Between Cooking Sessions and Orders')
plt.xlabel('Total Cooking Sessions')
plt.ylabel('Total Orders')
plt.show()
```

# Relationship Between Dish name, Duration and customer's Favorite Meal

```python
plt.figure(figsize=(12, 6), facecolor='lightblue')
sns.barplot(x='Dish Name', y='Duration (mins)', hue='Favorite Meal', data=full_data, palette='viridis')
plt.title('Cooking Duration for Each Dish by Favorite Meal')
plt.xlabel('Dish Name')
plt.ylabel('Duration (Minutes)')  
plt.legend(title='Favorite Meal')
plt.show()
```

# Relationship between orders per location 

```python
location_activity = full_data.groupby('Location').agg(
    TotalOrders=('Order ID', 'nunique')).reset_index()

plt.figure(figsize=(12, 6), facecolor='pink')
sns.barplot(x='Location', y='TotalOrders', data=location_activity, palette='coolwarm')
plt.title('Total Orders by Location')
plt.xlabel('Location')
plt.ylabel('Total Orders')
plt.show()
```
